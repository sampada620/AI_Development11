```python
import sqlite3

conn = sqlite3.connect(":memory:")  # In-memory database
cursor = conn.cursor()
```

```python
cursor.execute("""
CREATE TABLE student_code (
    id INTEGER PRIMARY KEY,
    student_name TEXT,
    code TEXT,
    score INTEGER
)
""")
```

```python
formatted_code = """
x = 10
print(x)
"""  # Example Python code as string

cursor.execute("""
INSERT INTO student_code (student_name, code, score)
VALUES (?, ?, ?)
""", ("Sampada", formatted_code, 65))

conn.commit()
```

```python
cursor.execute("SELECT * FROM student_code")
rows = cursor.fetchall()

for row in rows:
    print(row)
```
